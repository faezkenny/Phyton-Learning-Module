# Fibonacci Sequence Generator
This notebook generates the Fibonacci sequence up to a specified number of terms.

```python
def generate_fibonacci(n):
    sequence = []
    a, b = 0, 1
    for _ in range(n):
        sequence.append(a)
        a, b = b, a + b
    return sequence

n_terms = 10
fib_data = generate_fibonacci(n_terms)
print(f"First {n_terms} terms of the Fibonacci sequence:")
print(fib_data)
```
